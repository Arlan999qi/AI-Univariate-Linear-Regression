```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([2.1, 1.8, 6.1, 1.1, 4.0])
y_train = np.array([6.0, 5.1, 9.6, 1.9, 7.5])
```

```python
def compute_cost(x, y, w, b):
    cost = 0
    m = x_train.shape[0]
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i]) ** 2
    total_cost = 1 / (2 * m) * cost
    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    dj_dw = 0
    dj_db = 0
    m = x_train.shape[0]
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, numb_iter, cost_function, gradient_function):
    w = w_in
    b = b_in
    j_history = []
    p_history = []
    for i in range(numb_iter):
        dj_dw, dj_db = gradient_function(x, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        j_history.append(cost_function(x, y, w, b))
        p_history.append([w, b])
    return w, b, j_history, p_history
        
        
```

```python
w_init = 0
b_init = 0
alpha = 0.01
iterations = 1000
```

```python
def compute_model_output(x, w, b):
    m = x_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
pt_f_wb = compute_model_output(x_train, w_init, b_init)
plt.plot(x_train, pt_f_wb, c='b', label='Our prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual returns')
plt.title("Apple stock returns")
plt.ylabel("% change (Apple corp)")
plt.xlabel("% change (Market)")
plt.legend()
plt.show()
```

```python
w_final, b_final, j_history, p_history = gradient_descent(x_train, y_train, w_init, b_init, alpha, iterations, compute_cost, compute_gradient)
```
