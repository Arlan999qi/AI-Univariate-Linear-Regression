```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([50.0, 70.0, 90.0, 110.0, 130.0])
y_train = np.array([150.0, 200.0, 260.0, 320.0, 370.0])
print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

```python
m = x_train.shape[0]
print(f"The size of the training set is {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"x_{i} = {x_i}")
print(f"y_{i} = {y_i}")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (in meters square)")
plt.show()
```

```python
w = 2.85
b = 3
print(f"w = {w}")
print(f"b = {b}")
```

```python
def compute_model_output(x, w, b):
    m = x_train.shape[0]
    f_wb = np.zeros(x_train.shape[0])
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
pt_f_wb = compute_model_output(x_train, w, b)
plt.plot(x_train, pt_f_wb, c='b', label='Our prediction')
plt.scatter(x_train, y_train, marker='x', c="r", label='Actual prices')
plt.title("Housing prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (in meters square)")
plt.legend()
plt.show()
```
